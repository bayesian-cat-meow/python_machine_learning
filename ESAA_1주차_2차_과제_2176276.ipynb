# 개념 정리

Pandas


*   효율적인 데이터 처리를 가능케하는 파이썬 라이브러리






DataFrame


*   행과 열로 이루어진 2차원 데이터 구조체
*   index를 개별 데이터를 고유하게 식별하는 key로 사용


Series

*   열이 하나인 데이터 구조체
*   마찬가지로 index 사용


1. read_csv(), read_table(), read_fwf()

*   디렉토리 경로 설정
*   read_csv(): 콤마로 구분된 문자 기반의 파일 포맷을 데이터 프레임으로 불러옴
*   read_table(): 탭(\t)으로 구분된 문자 기반의 파일 포맷을 데이터 프레임으로 불러옴
*   read_fwf(): 고정 길이 기반의 칼럼 포뱃을 데이터 프레임로 불러옴


2. 불러온 데이터 확인



*   DataFrame.head(): 맨 앞에 있는 N개 행의 데이터를 보여줌
*   DataFrame.info(): 각 변수의 데이터 개수, 타입, 결측치 개수 등을 보여줌
*   Dataframe.describe(): 숫자형 변수에 대한 데이터 분포도를 보여줌
*   value_counts(): 해당 열의 값 유형과 건수를 보여줌




3. 데이터프레임 변환



*   ndarray로 변환하기: array = df.values
*   리스트로 변환하기: list = df.values.tolist()
*   딕셔너리로 변환하기: dict = df.to_dict()



4. DataFrame.drop()


*   labels: 삭제하고 싶은 칼럼명
*   axis: axis=0은 행 방향, axis=1은 열 방향
*   inplace: inplace=True는 반환값을 None으로 변환



5. [ ], iloc[ ], loc[ ]


*   reset_index(): 인덱스를 새로운 칼럼으로 추가
*   DataFrame['칼럼명']으로 칼럼 지정
*   iloc[  ]: 행과 열의 위치로 데이터 추출
*   loc[  ]: 명칭 기반으로 데이터 추출




6. 불린 인덱싱
*   조건을 이용해 데이터 필터링
*   and(&), or(|), not(~) 사용

7. 정렬, aggregation, GroupBy 사용

*   sort_values(): 데이터프레임을 유지한 채 정렬 결과 적용
*   aggregation 함수: min(), max(), sum(), count()
*   groupby(): 입력한 칼럼대로 groupby

8. isna(), fillna()
*   isna(): 결손 데이터(NaN)인지 아닌지를 True와 False로 출력
*   fillna(): 칼럼 내 결손 데이터를 입력한 데이터로 바꿈


9. apply lambda

* lambda 입력인자 : 계산식
* DataFrame.apply(lambda 입력인자 : 계산식)
* apply lambda 내 if, else 사용 가능

# DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환 (p.49)


넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

```python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print("1차원 리스트로 만든 DataFrame:\n", df_list1)

#넘파이 ndarray를 이용해 Dataframe 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print("1차원 ndarray로 만든 DataFrame:\n", df_array1)
```

```python
#3개의 칼럼명 필요함
col_name2 = ['col1', 'col2', 'col3']

#2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
        [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print("2차원 리스트로 만든 DataFrame:\n", df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print("2차원 ndarray로 만든 DataFrame:\n", df_array2)
```

# apply lambda 식으로 데이터 가공

```python
def get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))
```

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:',lambda_square(3))
```

```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' )
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else
                                                                                  'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'

    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
